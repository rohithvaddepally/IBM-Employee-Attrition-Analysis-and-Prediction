### 6.7 Feature Scaling or data encoding

Feature Scaling is a technique to standardize the independent features present in the data in a fixed range.

#### Label Encoding and Standard Scaling

```python
#importing label encoder and standard scaler libraries
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
le = LabelEncoder() #label encoder is stored in le
for feature in categorical_features:
    empatt[feature] = le.fit_transform(empatt[feature]) #fitting the le in Categorical features
```

```python
sc = StandardScaler() #standard scaler is stored in sc
for feature in numerical_features:
    empatt[feature] = sc.fit_transform(np.array(empatt[feature]).reshape(-1,1)) #fitting the sc in numerical features
```

```python
empatt['Attrition'] = le.fit_transform(empatt['Attrition']) #fitting the target column for le
```

### 6.8 Train and test data split

```python
X = empatt.drop("Attrition",axis=1) #independent variable 
y = empatt["Attrition"] #dependent variable
```

```python
#Importing train_test_split from sklearn library
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1) #Here data is split into 70% train data and 30% test data.
```

```python
#printing the shape of the variables X_test, y_test, X_test, y_test
print(X_train.shape)
print(y_train.shape)
print("-----------------------")
print(X_test.shape)
print(y_test.shape)
```

Before applying different machine learning algorithms to the data, We must ensure that target train data is balanced. To check that: 

```python
#printing all the unique values i.e; 0 and 1 in the y_train. Calling it as before oversampling 
print("Before OverSampling, counts of label '1': {}".format(sum(y_train == 1))) 
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train == 0))) 
```

The values in the target train data i.e; y_train should never be imbalanced. To resolve this  SMOTE technique should be used.

### 6.9 Resolving the imbalance problem

SMOTE: It is abbreviated as Synthetic Minority Oversampling Technique. It is basically used to resolve the imbalance problem in the data

```python
!pip install imblearn 
```

```python
#importing SMOTE from imblearn
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=0) #now stored in sm
oversampled_X_train, oversampled_y_train = sm.fit_resample(X_train, y_train) #fitting the original X_train,y_train and resampling it into new oversmapled_X_train and oversampled_y_train
```

```python
#printing the shape and size after the oversampling
print("After OverSampling, the shape of oversampled_X_train: {}".format(oversampled_X_train.shape))
print("After OverSampling, the shape of oversampled_y_train: {} \n".format(oversampled_y_train.shape))
```

```python
#printing the values i.e; 0 and 1 of the oversampling_y_train after the oversampling
print("After OverSampling, counts of label '1': {}".format(sum(oversampled_y_train == 1))) 
print("After OverSampling, counts of label '0': {}".format(sum(oversampled_y_train == 0))) 
```

Now, the oversampled_y_train data is balanced.

```python
plt.figure(figsize = (7, 7))
plt.pie([sum(oversampled_y_train == 1), sum(oversampled_y_train == 0)], labels=['YES', 'NO'], autopct='%1.0f%%', colors = ['lightblue','#66b3ff'])
plt.title("After resolving the imbalance problem in target training values", fontsize=20)
plt.show()
```

```python
logreg = LogisticRegression() #logisticregression is now stored in logreg
logreg.fit(oversampled_X_train,oversampled_y_train) #fitting the model
```

```python
print("Accuracy percent of the logistic Regression model: ", accuracy_score(y_test,log_y_pred)*100)
```

```python
print("Confusion Matrix : \n", confusion_matrix(y_test,log_y_pred))
```

```python
print("ROC AUC Score : ", roc_auc_score(y_test, log_y_pred))
```

```python
print("Classification Report: \n",  classification_report(y_test,log_y_pred))
```
